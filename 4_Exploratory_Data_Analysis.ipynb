##Reading external data from Kaggle

```python
import kagglehub
path = kagglehub.dataset_download("acadisheducation/concrete-dataset-linear-regression")

import pandas as pd
import os

# The 'path' variable contains the directory where the Kaggle dataset was downloaded.
# Let's list the contents of this directory to identify the CSV file.

print(f"Checking for CSV files in the Kaggle dataset directory: {path}")

files_in_directory = os.listdir(path)
csv_files = [f for f in files_in_directory if f.endswith('.csv')]

if len(csv_files) == 1:
    # If there's exactly one CSV file, assume it's the intended dataset and load it.
    single_csv_file_name = csv_files[0]
    full_file_path = os.path.join(path, single_csv_file_name)
    df = pd.read_csv(full_file_path)
    print(f"Successfully loaded '{single_csv_file_name}' into DataFrame 'df'.")
    print("Displaying the first 5 rows of the DataFrame:\n")
    display(df.head())
elif len(csv_files) > 1:
    print("Multiple CSV files found in the downloaded directory. Please specify which 'single dataset' you'd like to load:")
    for i, csv_file in enumerate(csv_files):
        print(f"{i+1}. {csv_file}")
    print("For example, if you want to load the first file, you can run:\n`df = pd.read_csv(os.path.join(path, csv_files[0]))`")
else:
    print("No CSV files found in the specified Kaggle dataset directory. Please ensure the download was successful and contains a CSV file.")
```

#Removing Outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set to store indices of rows that contain any outliers
outlier_indices = set()

print("Starting outlier detection and collecting outlier row indices...")

# Loop through each column in the DataFrame
for column in df.columns:
    # Only process numerical columns
    if df[column].dtype in ['int64', 'float64']:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Detect outliers for the current column
        current_column_outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        if not current_column_outliers.empty:
            # Add indices of rows with outliers to the set
            outlier_indices.update(current_column_outliers.index.tolist())
            print(f"Found {len(current_column_outliers)} outliers in column '{column}'.")

print(f"\nTotal unique rows identified with at least one outlier: {len(outlier_indices)}")

print("--- Creating df_cleaned by removing rows with outliers ---")
# Create df_cleaned by dropping rows identified as containing outliers
df_cleaned = df.drop(index=list(outlier_indices))

print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")
print("Displaying the first 5 rows of df_cleaned:")
display(df_cleaned.head())
```
