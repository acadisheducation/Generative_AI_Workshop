# Sample Code

```python


from google.colab import files
import pandas as pd
import io
from google.colab import runtime

print("Please select your dataset file to upload:")
uploaded_files = files.upload()

if uploaded_files:
    # Assuming the user uploads one primary file for analysis.
    # We'll take the first uploaded file's name.
    uploaded_file_name = list(uploaded_files.keys())[0]
    file_content = uploaded_files[uploaded_file_name]

    print(f"File '{uploaded_file_name}' uploaded successfully. Attempting to load into DataFrame 'df'...")

    try:
        # Attempt to infer file type and load accordingly
        if uploaded_file_name.lower().endswith('.csv'):
            df = pd.read_csv(io.BytesIO(file_content))
        elif uploaded_file_name.lower().endswith(('.xls', '.xlsx')):
            df = pd.read_excel(io.BytesIO(file_content))
        elif uploaded_file_name.lower().endswith('.json'):
            df = pd.read_json(io.BytesIO(file_content))
        else:
            print(f"Unsupported file format for automatic loading: {uploaded_file_name}. Please provide a CSV, Excel, or JSON file.")
            df = None # Set df to None to indicate failure to load

        if df is not None:
            print("Successfully loaded data into DataFrame 'df'.")
            print("Displaying the first 5 rows:")
            display(df.head())

    except Exception as e:
        print(f"An error occurred while loading the file: {e}")
        print("Please ensure the file is correctly formatted for the detected type.")
else:
    print("No file was uploaded. Please try again.")

#Ending the Runtime after Sample code has run
from google.colab import runtime
print("Task finished. Terminating Colab session.")
runtime.unassign()

```
